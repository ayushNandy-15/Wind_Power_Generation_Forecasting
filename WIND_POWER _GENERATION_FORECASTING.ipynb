```python
import pandas as pan
import numpy as np 
import matplotlib.pyplot as mat 
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler 
```

```python
location1 = pan.read_csv('Location1.csv')
location2 = pan.read_csv('Location2.csv')
location3 = pan.read_csv('Location3.csv')
location4 = pan.read_csv('Location4.csv')
```

```python
location1.head()
```

```python
location1['Location'] = 'Location1'
location2['Location'] = 'Location2'
location3['Location'] = 'Location3'
location4['Location'] = 'Location4'

# Concatenate to a dataframe
merged_data = pan.concat([location1, location2, location3, location4], ignore_index=True)

merged_data.head()
```

```python
merged_data.to_csv('merged_locations.csv', index=False)
```
